```python
df = pd.read_csv("..\Data\sales_marketing_data.csv")
df.head()
```

### Checking and cleaning Data

```python
df.columns
```

```python
df.rename(columns = {" Income ":"Income"}, inplace = True)
```

```python
df["Income"] = df["Income"].str.replace("$","",regex = True).str.replace(",","",regex = True)
df["Income"] = df["Income"].astype(float)
```

```python
# df["Dt_Customer"] = df["Dt_Customer"].dt.strftime(format = "%yyyy/%mm/%dd")
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"])
df["Dt_Customer"] = df["Dt_Customer"].dt.strftime("%d/%m/%y") #dtype cột này là object(string)
```

```python
df.shape
```

```python
df.isnull().sum() #Check null
```

```python
fig, ax = plt.subplots(figsize = (15,8))

sns.histplot(
    data = df["Income"],
    kde = True,
    color = "red"
)

plt.show()
```

+ Ở đây chúng ta thấy Data bị Right-skewed, nên chúng ta sẽ fill các giá trị null value bằng giá trị Median sẽ gần đúng nhất

```python
df["Income"].mean()
```

```python
df["Income"].fillna(value = df["Income"].mean(), inplace = True) #fill na
```

+ Ngoài ra chúng ta sẽ chia phân khúc khách hàng ra theo độ tuổi

```python
df.head()
```

```python
df["Age"] = 2021 - df["Year_Birth"] #Age of each customer
df["Age_Category"] = df["Age"].apply(lambda x: "Senior Citizen" if x >= 60 else "Adult" if x > 25 else "Youth" ) #Age per Category
df.head()
```

### ANALYZING

```python
df.head()
```

```python
df_cat = df.loc[:, df.dtypes == np.object]
df_cat.head()
```

```python
#Plotting
fig, ax = plt.subplots(1, 3,figsize = (22,6))

sns.countplot(
    data = df_cat,
    x = df_cat["Education"],
    ax = ax[0]
)

sns.countplot(
    data = df_cat,
    x = df_cat["Age_Category"],
    ax = ax[1]
)

sns.countplot(
    data = df_cat,
    x = df_cat["Marital_Status"],
    ax = ax[2]
)

#Decorate
ax[0].set_xlabel(None)
ax[0].set_ylabel(None)
ax[0].bar_label(ax[0].containers[0], padding = 2)

ax[1].set_xlabel(None)
ax[1].set_ylabel(None)
ax[1].bar_label(ax[1].containers[0], padding = 2)

ax[2].set_xlabel(None)
ax[2].set_ylabel(None)
ax[2].bar_label(ax[2].containers[0], padding = 2)

plt.show()
```

#### Insight
+ Ở đây chúng ta có thể thấy các khách hàng tại cửa hàng chủ yếu là các tập khách hàng có học thức cao, bằng cử nhân là chủ yếu, sau đó là Tiến sĩ.
+ Tập khách hàng chủ yếu từ 25 đến 60 tuổi là chủ yếu.
+ Chủ yếu khách hàng khi đến cửa hàng là các cặp đôi, cặp vợ chồng là nhiều nhất căn cứ theo 3 biểu đồ trên

```python
fig, ax = plt.subplots(figsize = (12,7))

sns.countplot(
    data = df_cat,
    x = df_cat["Country"]
)

#Decorating
ax.set_xlabel(None)

ax.bar_label(ax.containers[0], padding = 3)
plt.show()
```

+ Ta có thể thấy được rằng SP là nơi có lượng tiêu thụ hàng hóa cao nhất, cao hơn hẳn so với phần còn lại 

```python
amount = [col for col in df.columns if "Mnt" in col]
df[amount].sum()
```

```python
purchases = [col for col in df.columns if "Purchases" in col]
df[purchases].sum()
```

```python
fig, ax = plt.subplots(2, 1,figsize = (10,17))

df[amount].sum().plot.bar(ax = ax[0], color = "green")
df[purchases].sum().plot.bar(ax = ax[1], color = "lightblue")

#Decorating
ax[0].set_xticklabels(df[amount], rotation = 360)
ax[0].bar_label(ax[0].containers[0], padding = 2)
ax[1].set_xticklabels(df[purchases], rotation = 360)
ax[1].bar_label(ax[1].containers[0], padding = 2)
plt.show()
```

+ Mặt hàng bán chạy nhất của hệ thống cửa hàng là Rượu vang (Wines), tiếp theo sau đó là thịt (Meat)
+ Chủ yếu khách hàng sẽ đến trực tiếp cửa hàng để mua

=> Cửa hàng nên đẩy mạnh hơn việc sale các sản phẩm khác như Cá (Fish) và trái cây (Fruits).

=> Hệ thống cửa hàng nên phát triển thêm hệ thống bán hàng online để tăng thêm doanh thu.
