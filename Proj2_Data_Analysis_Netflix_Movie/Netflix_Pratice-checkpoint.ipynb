### Data Cleaning

- Drop duplicate value
- Drop show_id (seem useless for analyzing)

```python
data[data.duplicated(['title','country','type','release_year'])]
```

```python
dups = data.duplicated(['title','country','type','release_year'])
data[dups]
```

```python
data = data.drop_duplicates(['title','country','type','release_year'])
```

```python
data = data.drop('show_id', axis = 1 )
```

- Replace null values by Unknown

```python
null = data.isnull().sum()
null[null>0]
```

- All null values in 'director', 'country' and 'cast' will assume as Unknown

```python
data['director'] = data['director'].replace(np.nan, 'Unknown')
data['cast'] = data['cast'].replace(np.nan, 'Unknown')
data['country'] = data['country'].replace(np.nan, 'Unknown')
```

### EDA (Exploratory Data Analysis)

#### Now, it's time to understand some questions:
- We will understand and find what is the most rating on Netflix
- Comparing the frequency between Movie and TV Shows
- Top5 country produce Movie and TV Show

```python
# Prepare Plot Data
plot_data = data['rating'].value_counts(normalize = True).reset_index()
plot_data
```

```python
fig, ax = plt.subplots(figsize = (15,8))

plot_data.plot.bar(ax=ax)

ax.set_xticklabels(plot_data['index'], rotation = 'horizontal', fontweight = 'bold')
ax.set_xlabel('Rating', fontsize = 15)
ax.set_ylabel('Percentage', fontsize = 15)
ax.set_title('Distribution of rating Categories', fontsize = 20)
ax.set_ylim(0, 0.35)

ax.get_legend().remove()
plt.show()
```

- We can see obviously that TV-MA is the mode, TV-14 is second and go on

TV-MA: This program is specifically designed to be viewed by adults and therefore may be unsuitable for children under 17.

TV-14: This program contains some material that many parents would find unsuitable for children under 14 years of age.

- Which mean Netflix is focusing on Movie and TV shows for adults and mature customer segment

```python
fig,ax = plt.subplots(figsize = (10,10))

sns.countplot(
    x = 'rating',
    hue = 'type',
    data = data
)
ax.set_xticklabels(data['rating'], fontsize = 10)
ax.set_xlabel(None)
ax.set_ylim(0,1400)

plt.legend(loc = 'upper right', prop={'size':12})
plt.show()
```

- We can say that Movie is the majority category in rating category on Netflix
