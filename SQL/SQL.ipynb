```python
query = """
SELECT 
    first_name,
    last_name,
    email
FROM 
    customer 
        
LIMIT 5 
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT
    first_name,
    last_name,
    (first_name || '' || last_name) as full_name
FROM
    customer
LIMIT 5
"""
pd.read_sql(query,conn)
```

```python
query = """
select
    *
from film
where
    rental_duration >= 6 and
    rental_rate = 2.99  
limit 10
"""
pd.read_sql(query, conn)
```

```python
query = """
select *
from
    customer
where
    lower(first_name) in ('bob', 'rodriguez', 'adam')
limit 10
"""
pd.read_sql(query, conn)
```

```python
query = """
select *
from
    customer
where
    lower(first_name) like 'ann%'
limit 5
"""
pd.read_sql(query, conn)
```

```python
query = """
select *
from
    customer
where
    email like '%.org'
limit 5
"""
pd.read_sql(query,conn)
```

* we can use OFFSET to skip first line

```python
query = """
select
    title,
    rental_rate,
    length
from
    film
where
    length >= 100 and
    rental_rate <= 2
limit 10
offset 3 -- skip 3 giá trị đầu tiên 
"""

pd.read_sql(query, conn)
```

```python
query = """
select *
from
    payment
where
    payment_date >= '2007-02-15 00:00:00' and
    payment_date < '2007-02-16 00:00:00'
limit 5
"""
pd.read_sql(query, conn)
```

```python
query = """
select
    film_id,
    title,
    cast(rental_rate as integer) as rate_int --cast dùng để ép kiểu
from
    film    
limit 5

"""

pd.read_sql(query,conn)
```

```python
query = """
select
    f.film_id,
    title,
    l.name,
    c.name
from film as f
inner join language as l
on f.language_id = l.language_id
inner join film_category
on f.film_id = film_category.film_id
inner join category as c
on film_category.category_id = c.category_id
limit 10
"""

pd.read_sql(query, conn)
```

```python
query = """
select 
    c.customer_id,
    c.first_name,
    c.last_name,
    sum(p.amount) payment_per_id,
    count(*) num_payment,
    (sum(p.amount)/count(*)) as amount_per_transaction
from customer c
inner join payment p 
on c.customer_id = p.customer_id
group by c.customer_id, c.first_name, c.last_name
order by payment_per_id
"""

pd.read_sql(query, conn)
```
